```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt

boston_data = load_boston()
target = boston_data.target
df = pd.DataFrame(data=boston_data.data, columns=boston_data.feature_names)
print(df.head())
```

```python
crim_data = df['CRIM']
df_crim = pd.DataFrame(crim_data)
hist_crim = df_crim.hist(bins=20)
```

```python
age_data = df['AGE']
df_age = pd.DataFrame(age_data)
hist_age = df_age.hist(bins=20)
```

```python
crim_data_resaped = crim_data.values.reshape(-1, 1)
regression_crim = LinearRegression().fit(crim_data_resaped,
                                       target)
predict_crim = regression_crim.predict(crim_data_resaped)
plt.figure()
plt.scatter(x=crim_data, y=target)
plt.plot(crim_data, predict_crim, color='red')
plt.show()
```

```python
age_data_resaped = age_data.values.reshape(-1, 1)
regression_age = LinearRegression().fit(age_data_resaped,
                                       target)
predict_age = regression_age.predict(age_data_resaped)
plt.figure()
plt.scatter(x=age_data, y=target)
plt.plot(age_data, predict_age, color='red')
plt.show()
```

House price is strongly negatively dependant on crime rate and weakly on house age. Linear regression is not bad for age, but for sure is not the best choice for crime rate.
